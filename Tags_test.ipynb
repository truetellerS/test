##Import libraries and data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('dark_background')
table = pd.read_excel('/content/Статистика по тегам.xlsx')
df = table.copy()
new_header = df.iloc[0] 
df = df[1:] 
df.columns = new_header
df = df.rename(columns={col: col.replace('request_tags ', '') for col in df.columns})
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 18}
```

##Анализ 
Посчитаем количество уникальных тегов по каждому месяцу и наиболее часто встречающиеся теги во всех месяцах.

```python
plt.rc('font', **font)

unique_tags = df.apply(lambda x: len(x.unique()))
top_tags = df.stack().value_counts().head(15)

plt.figure(figsize=(22, 12))
unique_tags.plot(kind='bar', rot=0)
plt.title('Количество уникальных тегов по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Количество уникальных тегов')
plt.show()

plt.figure(figsize=(19, 12))
sns.barplot(x=top_tags.values, y=top_tags.index)
plt.title('Топ-15 наиболее часто встречающихся тегов')
plt.xlabel('Количество упоминаний')
plt.ylabel('Тег')
plt.show()
```

###Выведем топ 5 тегов по каждому месяцу

```python
for month in df.columns:
    tags_count = df[month].value_counts().head(5)
    plt.figure(figsize=(10, 8))
    plt.title(f'Top 5 tags for {month}')
    tags_count.plot(kind='bar')
    plt.xlabel('Tags')
    plt.ylabel('Count')
    plt.show()
```
